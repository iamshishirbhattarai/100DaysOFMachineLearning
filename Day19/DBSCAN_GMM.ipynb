## DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X,y = make_moons(n_samples=1000, noise=0.05)
dbscan = DBSCAN(eps=0.05, min_samples=5)
dbscan.fit(X)
dbscan.labels_
```

Those with the -1 labels are anomalies.

```python
dbscan.core_sample_indices_
```

The above model has a lot of clusters and many of the instances are considered anomalies. So, changing eps to 0.2

```python
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan.fit(X)
dbscan.labels_
```

### Gaussian Model Mixture

```python
from sklearn.mixture import GaussianMixture

gm = GaussianMixture(n_components=3, n_init=10)
gm.fit(X)
gm.weights_  #determines how much of the data is explained by each Gaussian (cluster)
```

```python
gm.means_  # Contains the center points (mean vectors) of each Gaussian component
```

```python
gm.covariances_    #describes the shape and spread of each Gaussian, showing how the data in each cluster is distributed.
```

```python
#checking whether algorithm is converged
gm.converged_
```

```python
#number of iterations it took
gm.n_iter_
```

```python
gm.predict(X)
```

### Bayesian Gaussian Mixture Model

Rather than manually finding the optimal number of clusters, use of this model provides us the weight equals or near to zero for the unnecessary clusters. Let us see this by fixing the number of clusters as 15 in the same problem as above.

```python
from sklearn.mixture import BayesianGaussianMixture

bgm = BayesianGaussianMixture(n_components=15, n_init=30, random_state=42)
bgm.fit(X)
bgm.weights_.round(2)
```

See the unnecessary clusters are provided 0 weights !!
